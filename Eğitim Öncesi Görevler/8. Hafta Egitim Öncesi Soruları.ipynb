# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle:
    def __init__(name,speed,mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
v1 = Vehicle("School Volvo", "180", "12")

vehicle = [v1]
for v in vehicle:
    print(f"{v.name}: {v.speed} {v.mileage})
````
